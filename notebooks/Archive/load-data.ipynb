```python
# Connect to the SQLite database
conn = sqlite3.connect(os.path.join(loading_dir, "journeys_nzta_20240820020304.db"))

# Create a cursor object
cursor = conn.cursor()

# List all tables in the database
cursor.execute("""
    SELECT name, sql 
    FROM sqlite_master 
    WHERE type='table'
""")

# Fetch all rows returned by the query
tables = cursor.fetchall()

# Print the table names
for table in tables:
    print(table[0])
```

```python
# Create a cursor object
cursor = conn.cursor()

# Specify the table name
table_name = 'cameras'

# Execute a PRAGMA statement to get the table structure
cursor.execute(f"""
    PRAGMA table_info([{table_name}])
""")

# Fetch all rows returned by the query
columns = cursor.fetchall()

# Print the column information
for column in columns:
    print(f"Column Name: {column[1]}")
    print(f"Data Type: {column[2]}")
    print(f"Not Null: {'Yes' if column[3] == 1 else 'No'}")
    print(f"Default Value: {column[4]}")
    print(f"Primary Key: {'Yes' if column[5] == 1 else 'No'}")
    print("---")
```
